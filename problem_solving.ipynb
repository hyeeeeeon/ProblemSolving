* preprocess: 데이터 셋 전처리 함수
* count_column: title과 text(본문)에서 가장 많이 나오는 수 20개 선정 함수

```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

stop_words = set(stopwords.words('english'))

def preprocess(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
  tokens = word_tokenize(text)

  words = [word for word in tokens if word not in stop_words and len(word) > 1]
  return words

def count_column(df, column):
  all_words = []
  for text in df[column].dropna():
    all_words.extend(preprocess(text))
  return Counter(all_words).most_common(20)

title_counts = count_column(fake_df, "title")
text_counts = count_column(fake_df, "text")

print("Top 20 words in title column")
print(title_counts)

print("Top 20 words in text column")
print(text_counts)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

fake_title = count_column(fake_df, "title")
fake_text = count_column(fake_df, "text")

def plot_top_words(word_counts, title):
    words, counts = zip(*word_counts)  # 단어, 등장 횟수 분리

    plt.figure(figsize=(10, 6))
    plt.barh(words, counts)
    plt.gca().invert_yaxis()  # 큰 수가 위에 오게
    plt.title(title, fontsize=14)
    plt.xlabel("Frequency")
    plt.tight_layout()
    plt.show()

# 예시: 위에서 얻은 결과로 시각화
plot_top_words(fake_title, "Top 20 words in title column")
plot_top_words(fake_text, "Top 20 words in text column")
```

```python
real_title = count_column(real_df, "title")
real_text = count_column(real_df, "text")

def plot_top_words(word_counts, title):
    words, counts = zip(*word_counts)  # 단어, 등장 횟수 분리

    plt.figure(figsize=(10, 6))
    plt.barh(words, counts)
    plt.gca().invert_yaxis()  # 큰 수가 위에 오게
    plt.title(title, fontsize=14)
    plt.xlabel("Frequency")
    plt.tight_layout()
    plt.show()

# 예시: 위에서 얻은 결과로 시각화
plot_top_words(real_title, "Top 20 words in title column")
plot_top_words(real_title, "Top 20 words in text column")
```

```python
def unique_words(count_a, count_b):
  dict_a = dict(count_a)
  dict_b = dict(count_b)

  unique_a = set(dict_a.keys()) - set(dict_b.keys())
  unique_b = set(dict_b.keys()) - set(dict_a.keys())

  return unique_a, unique_b

title_unique = unique_words(fake_title, real_title)
text_unique = unique_words(fake_text, real_text)

print("Unique words in title column")
print(title_unique)

print("Unique words in text column")
print(text_unique)
```
